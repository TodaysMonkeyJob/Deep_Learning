```python
# Vectorization
import numpy as np

a = np.array([1, 2, 3, 4, 5])
print(a)
```

```python
import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a, b)
toc = time.time()
print(c)
print("Vectorization time: " + str(1000*(toc - tic)) + "ms")
```

```python
c = 0
tic = time.time()
for i in range(1000000):
    c += a[i] * b[i]
toc = time.time()
print(c)
print("For loop time: " + str(1000*(toc - tic)) + "ms")
```

```python
# One of Example with numpy
import numpy as np

A = np.array([[56.0, 0.0, 4.4, 68.0],
             [1.2, 104.0, 52.0, 8.0],
             [1.8, 135.0, 99.0, 0.9]])
print(A)
```

```python
cel = A.sum(axis = 0)
print(cel)

percentages = 100 * A / cel.reshape(1, 4)
print(percentages)
```
